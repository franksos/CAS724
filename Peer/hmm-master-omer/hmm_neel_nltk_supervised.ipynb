```python
import pickle
import nltk

neel_train_data_file = '../src/clean_data/neel_train_v1.p'
with open(neel_train_data_file, 'rb') as f:
    neel_train_data = pickle.load(f)
    
train_words = []
train_labels = []
train_labelled_words = []
for key in neel_train_data.keys():
    temp_words = neel_train_data[key]['words']
    temp_labels = neel_train_data[key]['labels']

    if len(temp_words) != len(temp_labels):
        print("error")
    
    temp_labelled_words = []
    for word, label in zip(temp_words, temp_labels):
        #if label != 'O':
        train_words.append(word)
        train_labels.append(label)
        temp_labelled_words.append((word,label))
    train_labelled_words.append(temp_labelled_words)
    '''
    for i in range(len(temp_labels)):
        if temp_labels[i] is not 'O':   
            words += temp_words[i]
            labels += temp_labels[i]
    '''
    
print(len(train_words) == len(train_labels))
print(len(train_words))
print(len(train_labelled_words))
print(train_words[0])
train_labelled_words[0]
```

```python
labels_set = list(set(train_labels))
labels_set
```

```python
words_set = set()

for word in train_words:
    for char in word:
        words_set.add(char)
    words_set.add(word)
words_set = list(words_set)
words_set
```

```python
print(len(train_words))
print(len(train_labelled_words))
train_labelled_words[0]
```

```python
org = 0
loc = 0
misc = 0
per = 0
o = 0
for l in train_labels:
    if l == 'ORG':
        org += 1
    elif l == 'LOC':
        loc += 1
    elif l == 'MISC':
        misc += 1
    elif l == 'PER':
        per += 1
    elif l == 'O':
        o += 1
print("org: ", org, "\tloc: ", loc, "\tmisc: ", misc, "\tper: ", per, "\tO: ", o)
```

```python
trainer = nltk.tag.hmm.HiddenMarkovModelTrainer(states=labels_set, symbols=words_set)
m = trainer.train_supervised(train_labelled_words)
```
